```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movies_df = pd.read_csv('./imdb-5000-movie-dataset/movie_metadata.csv')
movies_df.head()
```

```python
movies_df['movieIndex'] = movies_df.index
movies_df = movies_df[['movieIndex','movie_title','genres','imdb_score','movie_facebook_likes','content_rating','budget','gross']]
movies_df.head()
```

```python
movies_df['genres'] = movies_df.genres.str.split('|') 
movies_df.head()
```

```python
movieGenres_df = movies_df[['movieIndex','movie_title','genres']]
for index, row in movies_df.iterrows():
    for genre in row['genres']:
        movieGenres_df.at[index, genre] = 1
movieGenres_df = movieGenres_df.fillna(0)
movieGenres_df = movieGenres_df.drop('genres',1)
movieGenres_df.head()
```

```python
# n = int(input("Enter no of movies"))
# arr = []
# for i in range(n):
#     obj = {}
#     obj['movie_title'] = input("Enter movie title")
#     obj['rating'] = float(input("Enter rating"))
#     arr.append(obj)
# arr
userInputMovies = [{'movieIndex': 2, 'rating': 8},
              {'movieIndex': 7, 'rating': 6},
              {'movieIndex': 19, 'rating': 7},
              {'movieIndex': 57, 'rating': 9},
              {'movieIndex': 90, 'rating': 5},
              {'movieIndex': 80, 'rating': 6},
              {'movieIndex': 147, 'rating': 3},
              {'movieIndex': 89, 'rating': 9},
              {'movieIndex': 876, 'rating': 7},
              {'movieIndex': 567, 'rating': 9}]
```

```python
inputMovies = pd.DataFrame(userInputMovies)
inputMovies.head()
```

```python
userMovies = movieGenres_df[movieGenres_df['movieIndex'].isin(inputMovies['movieIndex'].tolist())]
userGenreTable = userMovies.drop('movieIndex',1).drop('movie_title',1)
userGenreTable = userGenreTable.reset_index(drop=True)
userGenreTable
```

```python
userProfile = userGenreTable.transpose().dot(inputMovies['rating'])
userProfile
```

```python
genreTable = movieGenres_df.set_index(movieGenres_df['movieIndex'])
genreTable = movieGenres_df.drop('movieIndex',1).drop('movie_title',1)
genreTable.head()
```

```python
recommendationTable_df = ((genreTable*userProfile).sum(axis=1))/(userProfile.sum())
recommendationTable_df = recommendationTable_df.sort_values(ascending=False)
recommendationTable_df.head()
```

```python
recommendations = movies_df[movies_df['movieIndex'].isin(recommendationTable_df.head(20).keys())]
recommendations
```

```python
highest_imdb = movies_df.sort_values(by='imdb_score',ascending=False)[['movie_title','imdb_score']].dropna()
highest10_imdb = highest_imdb.reset_index(drop=True).head(10)
highest10_imdb
```

```python
highest10_imdb.iloc[::-1].plot.barh(x='movie_title',color='orange' ,rot=0, figsize=(13,10), fontsize=12, legend=False)
plt.ylabel('Movies',fontsize='30')
plt.xlabel('IMDB Rating',fontsize='30')
plt.show()
```

```python
highest_fb = movies_df.sort_values(by='movie_facebook_likes',ascending=False)[['movie_title','movie_facebook_likes']].dropna()
highest10_fb = highest_fb.reset_index(drop=True).head(10)
highest10_fb
```

```python
highest10_fb.iloc[::-1].plot.barh(x='movie_title', rot=0, figsize=(13,10), fontsize=12, legend=False)
plt.ylabel('Movies',fontsize='30')
plt.xlabel('Facebook Likes',fontsize='30')
plt.show()
```

```python
by_content_rating = movies_df[['movieIndex', 'content_rating']].groupby('content_rating', axis=0).count()
by_content_rating.rename({'movieIndex': 'count'}, axis=1, inplace=True)
by_content_rating = by_content_rating.sort_values(by="count", ascending=False)
by_content_rating.head()
```

```python
colors_list = ['#003f5c','#ffa600','#58508d', '#bc5090','#ff6361']
by_content_rating.head()['count'].plot(kind='pie', figsize=(15, 6), autopct='%1.1f%%', startangle=90, shadow=True, labels=None, pctdistance=1.12, colors=colors_list)
plt.title('Proportion of movies by content rating',y=1.12) 
plt.axis('equal') 
plt.legend(labels=by_content_rating.index, loc='upper right') 
plt.show()
```

```python
gross_vs_budget = movies_df[['movie_title','gross','budget']].sort_values(by="gross",ascending=False)
gross_vs_budget.drop_duplicates(keep='first',inplace=True)
gross_vs_budget.head(10)
```

```python
barWidth = 0.4
r1 = np.arange(10)
r2 = [x + barWidth for x in r1]
plt.figure(figsize=(11, 10))
plt.barh(r1, gross_vs_budget['budget'].head(10), color='red', height=barWidth, edgecolor='white', align='edge', label='budget')
plt.barh(r2, gross_vs_budget['gross'].head(10), color='green', height=barWidth, edgecolor='white', align='edge', label='gross')
plt.gca().invert_yaxis()
ax = plt.gca()
ax.get_xaxis().get_major_formatter().set_scientific(False)
plt.title('Gross v/s Budget of highest grossing movies')
plt.xlabel('Amount in dollars($)', fontweight='bold')
plt.xticks(rotation = 30)
plt.ylabel('Movies', fontweight='bold')
plt.yticks([r + barWidth for r in range(10)], gross_vs_budget['movie_title'].head(10))
plt.legend()
plt.show()
```
